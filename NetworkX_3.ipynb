# Домашнее задание по библиотеке NetworkX

```python
import networkx as nx

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

## Задание 1 (1 балл)
Создайте граф и добавьте в него $5$ вершин с городами России, у каждой должен быть атрибут (характеристика) `time_diff`, означающий временную разницу с Москвой в часах.

```python
fly_graph = nx.Graph()

fly_graph.add_node('Ставрополь', time_diff = 7)
fly_graph.add_node('Краснодар', time_diff = 6)
fly_graph.add_node('Армавир', time_diff = 3)
fly_graph.add_node('Сочи', time_diff = 1)
fly_graph.add_node('Ейск', time_diff = 0)
```

## Задание 2 (1 балл)
Выберите какую-нибудь конкретную авиакомпанию (Аэрофлот, S7, Utair, etc).

В созданный граф добавьте рёбра между выбранными $5$ю городами, между которыми есть авиарейсы данной компании. У каждого ребра должны быть следующие атрибуты:
- расстояние между городами в километрах `distance`
- время перелёта в минутах `time`

При желании можно добавить в граф также несколько городов, являющихся пересадочными пунктами.

```python
#Компания "Аэрофлот"
fly_graph.add_edge('Краснодар', 'Ейск', distance = 13422, time = 5495)
fly_graph.add_edge('Краснодар', 'Сочи', distance = 3120, time = 65)
fly_graph.add_edge('Краснодар', 'Ставрополь', distance = 4206, time = 1350)
fly_graph.add_edge('Краснодар', 'Армавир', distance =  1313, time =   1220)
fly_graph.add_edge('Ставрополь', 'Армавир', distance = 3120+913, time = 1240)
fly_graph.add_edge('Ставрополь', 'Ейск', distance = 6120+3422, time = 1245)
fly_graph.add_edge('Армавир', 'Сочи', distance = 4206+913, time = 1195)
fly_graph.add_edge('Ейск', 'Сочи', distance = 11422+913, time = 2845)
fly_graph.add_edge('Армавир', 'Ейск', distance =  3422, time =  495)
fly_graph.add_edge('Ставрополь', 'Сочи', distance = 3120, time = 1245)
plt.figure(figsize=(20, 20)) 
nx.draw_networkx(fly_graph, node_color='green')
```

## Задание 3 (1 балл)
Выберите любую пару городов и постройте между ними следующие кратчайшие пути:
- без взвешивания
- со взвешиванием по расстоянию
- со взвешиванием по времени

```python
print(nx.shortest_path(fly_graph, 'Ставрополь', 'Сочи'))
print(nx.shortest_path(fly_graph, 'Ставрополь', 'Сочи', weight='distance'))
print(nx.shortest_path(fly_graph, 'Ставрополь', 'Сочи', weight='time'))
```
